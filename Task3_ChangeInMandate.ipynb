```python
import pandas as pd
fn = "newmandate_acc.xlsx"
fntest = "newmandate.xlsx"
dfTask2 = pd.read_excel("task1n2_PL_R2_v2.xlsx",sheetname= "Task2_Table", converters={'SI':str})
```

```python
dfADA = pd.read_excel(fn, converters= {'SFX':str})
dfADA['Acc Number'] = dfADA['Acc Number'].str.strip()
dfADA['Acc Number'].tolist()
dfTest =pd.read_excel(fntest,converters={'SFX':str, 'Corp CIN':str})
dfTest
```

```python
'''
output of the dfconcat, "x" is the change case "y" is the original credentials
'''
dfconcat = dfTest.merge(dfADA, left_on = ['ind CIN'], right_on= ['ind CIN'],
                       how = 'right')
dfconcat = dfconcat.drop(['Acc Number_x'],axis =1)
dfconcat = dfconcat.astype(str)
# type(dfconcat['Corp CIN_y'][0])
dfconcat
```

```python
dfTask2 = dfTask2.astype(str)
dfADA = dfADA.astype(str)
dfconcat = dfconcat.astype(str)
type(dfconcat['Corp CIN_x'][0])
```

```python
for row in dfconcat.iterrows():
    #print(pd.isnull(row[1]['SFX_x']))
    print(row[1]['SFX_x']=="nan")
```

```python
'''
these cases covers the column of remarks for the case if 'Corp CIN'
'ind CIN', 'SFX' and 'Name' columns are identical. Or else we go to the next section to append
the remarks. These for grouping cases

See the test cases in the 'Task3.jpg'
'''
testlist = []
NRIC = []

for row in dfconcat.iterrows():
    '''
    These are for "signer" cases of signer in D(beginning) not in F (Final)
    '''
    if (row[1]['SFX_y']=="nan") and (row[1]['Name_y']=="nan"):
        a = "Delete"
        testlist.append(a)
        NRIC.append(row[1]['ind CIN'])
    
    # credentials for  4 columns the same  
    elif (row[1]['Corp CIN_x']== row[1]['Corp CIN_y']) and (row[1]['SFX_x'] == row[1]['SFX_y']) and \
    (row[1]['Name_x']==row[1]['Name_y']):
        if row[1]['Group_x']!= "" and (row[1]['Group_y']=="nan"):
            a = "No Group"
            NRIC.append(row[1]['ind CIN'])
            testlist.append(a)
        elif (row[1]['Group_x']=="nan") and (row[1]['Group_y']!=""):
            a = "Grouping required in {}".format(row[1]['Group_y'])            
            NRIC.append(row[1]['ind CIN'])
            testlist.append(a)
        elif (row[1]['Group_x']!= "" and row[1]['Group_y']!= ""):
            a = 'regroup to {}'.format(row[1]['Group_y'])
            testlist.append(a)
            NRIC.append(row[1]['ind CIN'])
        elif (row[1]['Group_x']== row[1]['Group_y']) and (row[1]['Group_x']!=""):
            a ="No Change"
            testlist.append(a)
            NRIC.append(row[1]['ind CIN'])
        else:
            a = "please check"
            testlist.append(a)
            NRIC.append(row[1]['ind CIN'])
    elif (row[1]['SFX_x']!= row[1]['SFX_y']) and (row[1]['Name_x']!= row[1]['Name_y']):
        a="Add {} grouping".format(row[1].Group_y)
        testlist.append(a)
        NRIC.append(row[1]['ind CIN'])
    else:
        a= "Scenario recommendation not provided"
        testlist.append(a)
        NRIC.append(row[1]['ind CIN'])
print(len(testlist))
print(len(NRIC))

testlist_ic = list(set(zip(NRIC,testlist)))
print(testlist_ic)
```

```python
# since we already created remarks column back in "testlist_ic"
# make that column into a dataframe before merge with dfFinal1
NRIC = []
remarks = []
for item in testlist_ic:
    NRIC.append(item[0])
    remarks.append(item[1])
print(NRIC)
print(remarks)

# make into a dataframe
d = {'NRIC':NRIC, 'Remarks':remarks}
dfRemarks = pd.DataFrame(data=d)
print(dfRemarks)

# merge dfRemarks and dfconcat
dfRemarks1 = dfRemarks.merge(dfconcat, left_on = ['NRIC'], right_on= ['ind CIN'],how = 'right')
dfRemarks1
```

```python
# to make sure is unique individual, we append the "ind CIN", "SFX" ,and "Name" and then do a composite key
dfRemarks1['appendind'] = dfRemarks1['ind CIN'].map(str) + "_" + dfRemarks1['SFX_y'] + "_" + dfRemarks1['Name_y']

wordlist = dfRemarks1['appendind'].tolist()
wordfreq = []
for w in wordlist:
    wordfreq.append(wordlist.count(w))
pairs = list((zip(wordlist,wordfreq)))
pairs
```

```python
'''
Put the unique individuals and the frequency count together
Next, unsplit the key by "_"
Put each of the components back into columns
'''
pairs = list(set(pairs)) # unique pairs
print("Unique pairs: \n {}".format(pairs))

indIC = []
SFInd = []
NameInd = []
Frequency = []
for item in pairs:
    IC,SF,name = item[0].split("_")
    indIC.append(IC)
    SFInd.append(SF)
    NameInd.append(name)
    Frequency.append(item[1])
print(indIC)
```

```python
# create a dataframe based on the list above
d = {'ind CIN':indIC, 'SFX':SFInd, 'Name':NameInd, 'Count':Frequency}
df = pd.DataFrame(data=d)
```

```python
# now merge df with dfconcat by the field "ind CIN"
dfFinal = dfRemarks1.merge(df,left_on=['ind CIN'], right_on=['ind CIN'], how = 'left')
#print(dfFinal)
dfFinal1 = dfFinal[['ind CIN','Corp CIN_x','SFX', 'Name', 'Count']] # select needed columns
dfFinal1
```

```python
'''
Save the tablesheets in xlsx
'''
filename = 'task3.xlsx'
writer = pd.ExcelWriter(filename)

# write to csv the output
dfFinal1.to_excel(writer,sheet_name = "Task3_Table")
writer.save()
```
